```python
# This code was written by Daniel Schacher and its purpose is to simulate two interacting einstein solids
# Ver1.0 last edited: Sept-09-2018
from math import *
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
mpld3.enable_notebook()
```

```python
# finds the multiplicity of an einstein solid
def einsteinsolid(N , q):
    return factorial(q+N-1)/(factorial(q)*factorial(N-1))
```

```python
na = np.zeros((101, 5))  # array to hold the values for the first set of solids
na2 = np.zeros((101, 5)) # array to hold the values fot the second set of solids
# This loop will fill the arrays 
for q in range(0,101):
    na[q , 0] = q
    na[q , 1] = einsteinsolid(300 , q)
    na[q , 2] = 100 - q
    na[q , 3] = einsteinsolid(200 , 100 - q)
    na[q , 4] = (na[q , 3]) * (na[q , 1])
    na2[q , 0] = q
    na2[q , 1] = einsteinsolid(3000 , q)
    na2[q , 2] = 100 - q
    na2[q , 3] = einsteinsolid(2000 , 100 - q)
    na2[q , 4] = (na2[q , 3]) * (na2[q , 1])
# Data is placed into dataframe with labeled columns and indicies    
df = pd.DataFrame(data = na , columns = ['q(A)' , 'O(A)' , 'q(B)' , 'O(B)' , 'O(total)'])
df2 = pd.DataFrame(data = na2 , columns = ['q(A)' , 'O(A)' , 'q(B)' , 'O(B)' , 'O(total)'])
print(df, '\n', df2)    
```

```python
# plot of first set of solids where Na = 300 and Nb = 200
y = df['O(total)']
x = df['q(A)']
fig = plt.plot(x , y)
```

```python
# plot for second set of solids where Na = 3000 and Nb = 2000
y1 = df2['O(total)']
x1 = df2['q(A)']
fig = plt.plot(x1 , y1)
```

```python
# the plots describe a Gaussian distribution and as N increases the peak of the graph grows more narrow.
# the width of the peak is inversly proporsional to the root of N and proportional to q.
width1 = 100 / sqrt(300)
width2 = 100 / sqrt(3000)
print(width1 , width2)
# This means the probabilty density becomes more concentrated toward the center (q/2) as N grows.
```

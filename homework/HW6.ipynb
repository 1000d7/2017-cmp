```python
# Daniel Schacher 
# homework for section 9

from scipy.fftpack import fft2, ifft2
import numpy as np
import matplotlib.pyplot as plt

# Load data into an array
data = np.loadtxt("circular.txt",float)
plt.figure(0)
plt.imshow(data)
plt.title('Original data')
plt.savefig('data.png')

# fft on data then frequency density plot
fftdata = fft2(data)
plt.figure(1)
plt.imshow(np.log(np.abs(fftdata)))
plt.title('fft data')
plt.savefig('fftdata.png')

# fillter out high frequencies
r, c = fftdata.shape
keep_fraction = 0.1 # fraction of data to keep
fftdata[int(r*keep_fraction):int(r*(1-keep_fraction))] = 0
fftdata[:, int(c*keep_fraction):int(c*(1-keep_fraction))] = 0
plt.figure(2)
plt.imshow(np.log(np.abs(fftdata)))
plt.title('compressed data')
plt.savefig('compresseddata.png')

# inverse fft then plot real part
data2 = ifft2(fftdata)
data2 = (data2.real)
plt.figure(3)
plt.imshow(data2)
plt.title('unconpressed data')
plt.savefig('uncompresseddata.png')


```

The saved .png of the data is compressed from about 44kB to 8kB.
The resulting uncomptessed image retains most of the quality.
Next I'll compress a meme.

```python

from matplotlib import colors
cmap = colors.ListedColormap(['black','white'])

# Load data into an array
data = im = plt.imread('meme.png').astype(float)
plt.figure(0)
plt.imshow(data, cmap = cmap)
plt.title('Original data')
plt.savefig('data.png')

# fft on data then frequency density plot
fftdata = fft2(data)
plt.figure(1)
plt.imshow(np.log(np.abs(fftdata)), cmap = cmap)
plt.title('fft data')
plt.savefig('fftdata.png')

# fillter out high frequencies
r, c = fftdata.shape
keep_fraction = 0.3 # fraction of data to keep
fftdata[int(r*keep_fraction):int(r*(1-keep_fraction))] = 0
fftdata[:, int(c*keep_fraction):int(c*(1-keep_fraction))] = 0
plt.figure(2)
plt.imshow(np.log(np.abs(fftdata)), cmap = cmap)
plt.title('compressed data')
plt.savefig('compresseddata.png')

# inverse fft then plot real part
data2 = ifft2(fftdata)
data2 = (data2.real)
plt.figure(3)
plt.imshow((data2), cmap = cmap)
plt.title('unconpressed data')
plt.savefig('uncompresseddata.png')
```

The meme was compressed and then uncompressed with out any noticable loss in quality.

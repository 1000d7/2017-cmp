```python

# exponential distribution of random numbers
import numpy as np
import matplotlib.pyplot as plt

tau = 3.053*60

N_atom = 1000   
N_time = 1000   
mu = np.log(2)/tau
nud = [] # non uniform distribution list of numbers
for i in range(N_atom):    
    z0 = np.random.random()
    x0 = int(-np.log(1-z0)/mu) #the time to decay from random number
    nud.append(x0) # append new output to list

plt.hist(nud,100, density=1)
plt.xlim(0,1000)
plt.show()
print(nud)
```
